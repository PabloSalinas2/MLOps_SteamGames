```python
import pandas as pd
import gzip
import json
import ast
# crear una funcion para descomprimir el archivo 
path_games= '../PROYECTO INDIVIDUAL/Datos del Proyecto/steam_games.json.gz'
path_reviews='../PROYECTO INDIVIDUAL/Datos del Proyecto/user_reviews.json/australian_user_reviews.json'
path_items='../PROYECTO INDIVIDUAL/Datos del Proyecto/users_items.json/australian_users_items.json'
```

```python
# Leer el archivo steam_games

def descomprimir_archivo(archivo):
    try:
        with gzip.open(archivo,'rt') as archivo_descomprimido:
            datos=pd.read_json(archivo_descomprimido,lines=True)
        return datos
    except Exception as e:
        print('Error al descomprimir el archivo :', e)
```

```python
steam_games=descomprimir_archivo(path_games)
```

```python
# crear funcion para reemplazar los valores Free por 0 
def reemplazo_free_por_0(valor):
        if type(valor)==str:
            if "Free" in valor:
                return 0
            else: 
                return valor
        else:
             return valor
```

```python
import re

def extaer_año(cadena):
    patron=r"\d{4}"
    match=re.search(patron,cadena)
    if match:
        return match.group()
    else:
        return cadena
```

```python
df5=steam_games_listo['año_lanzamiento'].apply(extaer_año)
```

```python
df5
```

```python
# EXPLORARA VALORES NO NUMERICOS
import re
lista_num=[]
lista_sin_formato=[]
for i in df5.unique():
    for letra in i:
        patron1=r"\d"
        match1=re.search(patron1,i)
        if match1:
            lista_num.append(i)
        else:
            lista_sin_formato.append(i)
```
