```python
# CARGAR en un dataframe el archivo path_reviews 
data_list=[]
non_read=[]
with open(path_reviews,'r',encoding='utf-8') as file:  
    for number_line, line in enumerate (file,start=1):  # por que start 1 ? 
        try:
            data=json.loads(line)
            data_list.append(data)
        except json.JSONDecodeError as e:
            try:
                data=ast.literal_eval(line)
                if isinstance(data,dict):  # que hace isinstance ? 
                        data_list.append(data)
                else:
                     non_read.append(number_line)
            except json.JSONDecodeError as e:
                    non_read.append(number_line)
    
```

```python
# CREAR FUNCION PARA DESANIDAR ######### EXTRA QUE NO VA EN EL MODELO PERO REVISARLA A MODO DE ESTUDIO #############

def desanidar(diccionario_anidado,prefijo=''):
    diccionario_desanidado={}
    print('paso por aqui')
    for clave,valor in diccionario_anidado.items():
        print('iteracion',clave,'-',valor)
        if isinstance(valor,dict):
            diccionario_desanidado.update(desanidar(valor,prefijo+clave+'-'))
        else:
            diccionario_desanidado[prefijo+clave]=valor
        print(diccionario_desanidado)
    return diccionario_desanidado
```

```python
# CARGAR en un dataframe el archivo path_items
items_list=[]
no_leidos=[]
with open(path_items,'r',encoding='utf-8') as archivo:
    for nro_fila,fila in enumerate(archivo,start=1):
        try:
            datos=ast.literal_eval(fila)
            if isinstance(datos,dict):
                items_list.append(datos)
            else:
                no_leidos.append[nro_fila]
        except json.JSONDecodeError as e:
            no_leidos.append(nro_fila)
          

    
```

```python
# Eliminacion de columnas
steam_games=steam_games.drop(['url','reviews_url'],axis=1)
```

```python
# Reemplazar nulos por 0
steam_games['price']=steam_games['price'].fillna(0)
```

```python
# Cambiar tipos 
   # Reemplzar los str por valores numericos
no_numericos=[]
for valor in steam_games['price'].unique():
   if type(valor)!=float:
      print(valor,type(valor))
```

```python
# Si el elemento de la serie contiene la palabra "Free", reemplazarlo por 0 

# Reemplazar Nulos


def reemplazo_free_por_0(valor):
    try:
        if "Free" in valor:
            return 0
        elif 'Starting at $449.00' in valor:
            return 499.0
        else: 
            return valor
    except TypeError as e:
        return 0
```

```python
steam_games['price']=steam_games['price'].apply(reemplazo_free_por_0)
```
