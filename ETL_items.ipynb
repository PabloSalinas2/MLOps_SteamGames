##### ***1) LECTURA DE ARCHIVOS***

```python
import pandas as pd
import gzip
import json
import ast

path_items='../PROYECTO INDIVIDUAL/Datos del Proyecto/users_items.json/australian_users_items.json'

```

```python
items_list=[]
no_leidos=[]
with open(path_items,'r',encoding='utf-8') as archivo:
    for nro_fila,fila in enumerate(archivo,start=1):
        try:
            datos=ast.literal_eval(fila)
            if isinstance(datos,dict):
                items_list.append(datos)
            else:
                no_leidos.append[nro_fila]
        except json.JSONDecodeError as e:
            no_leidos.append(nro_fila)
          
```

##### ***2) CONSTRUCCION DE ETL***

```python
# DESANIDAR COLUMNA "items"

lista_desanidada_items=[]
for indice,diccionario in enumerate(items_list):
    for inidex,diccionario_sec in enumerate(diccionario['items']):
        lista_desanidada_items.append({'user_id':diccionario['user_id'],
                        'items_count':diccionario['items_count'],
                        'steam_id':diccionario['steam_id'],
                        'user_url':diccionario['user_url'],
                        'item_id':diccionario_sec['item_id'],
                        'item_name':diccionario_sec['item_name'],
                        'playtime_forever':diccionario_sec['playtime_forever'],
                        'playtime_2weeks':diccionario_sec['playtime_2weeks']
                        })

users_items=pd.DataFrame(lista_desanidada_items)
```

```python
# ELIMIACION de columnas innecesarias 

users_items.drop(['user_url'],axis=1,inplace=True)
```

```python
users_items.drop(['steam_id'],axis=1,inplace=True)
```

```python
users_items.drop(['items_count'],axis=1,inplace=True)
```

```python
# renombrar columnas para mejor identificacion 

users_items=users_items.rename(columns={'item_id':'id_juego','item_name':'nombre_juego','playtime_forever':'timpo_total_jugado','playtime_2weeks':'tiempo_jugado_ultimas_2semanas'})
```

```python
# Ver si hay valores nulos 

for columna in users_items.columns:

    nulo=users_items[columna].isnull().sum()
    print(f'{columna} valores nulos= ',nulo)
```
