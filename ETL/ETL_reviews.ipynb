#### **Librerias a utilizar**

```python
import pandas as pd
import gzip
import json
import ast
from textblob import TextBlob

```

#### **Extracción de datos**

```python
# Lee el archivo Json linea por linea y devuelve una lista de diccionarios

ruta='../Data Sets/user_reviews.json.gz'

reviews_list=[] # lista donde se almacenan los datos 

with gzip.open(ruta,'rt',encoding='utf-8') as archivo:
    for fila in archivo.readlines():
        datos=ast.literal_eval(fila)
        reviews_list.append(datos)


# Crea un DataFrame a partir de la lista de diccionarios obtenidas en el bloque de codigo anterior
user_reviews=pd.DataFrame(reviews_list)
user_reviews
```

#### **Desanidar Datos**

```python
user_reviews.shape
```

```python
# Exploracion 

user_reviews=user_reviews.explode('reviews').reset_index(drop=True) # desanidacion parcial de la columna reviews, devuelve una fila por cada diccionario contenido dentro de la fila
user_reviews

```

```python
# Normaliza columna "reviews"
datos_normalizados=pd.json_normalize(user_reviews['reviews'])
datos_normalizados
```

```python
# concatena el DataFrame con los datos normalizados 

user_reviews=pd.concat([user_reviews,datos_normalizados],axis=1)
user_reviews
```

```python
# Elimino columna "reviews"
user_reviews.drop('reviews',axis=1,inplace=True)
user_reviews.head(5)
```

La columna "posted" contiene la fecha en la que el usuario realizó una review, es necesario que los datos de esta columna esten en formato fecha, para ello es necesario hacer algunas transformaciones, ya que los datos están sucios, además a partir de esta columna añadiré una columna donde solo muestre el año.

```python
# Eliminar la Palabra "Posted" de la columna  , para ello separ en otra columna la palabra Posted y despues eliminarla 

user_reviews[['Posted','review_posted']]=user_reviews['posted'].str.split('Posted',expand=True)
user_reviews.head(5)
```

```python
# Elimina columnas: "posted" , "Posted" 

user_reviews.drop(['Posted','posted'],axis=1,inplace=True)
user_reviews.head(5)
```

```python
#Elimina el punto "." de la columna review_posted
user_reviews['review_posted']=user_reviews.review_posted.str.replace('.','')
user_reviews.head(5)
```

```python
# Eliminar espacioes en blanco 
user_reviews['review_posted']=user_reviews['review_posted'].str.strip()
```

```python
# cambiar columna a tipo fecha, aquellos campos con formato incorrecto son transformados en valores nulos 
user_reviews['review_posted']=pd.to_datetime(user_reviews['review_posted'],errors='coerce')
```

```python
# Eliminar registros donde la columna "review_posted" contenga valores nulos 
user_reviews=user_reviews.dropna(subset='review_posted').reset_index(drop=True)
user_reviews
```

```python
# Crear columna "year_review_posted"

user_reviews['year_review_posted']=user_reviews['review_posted'].dt.year
user_reviews
```

#### **Columna "item_id"**

```python
# cambia a tipo int

user_reviews['item_id']=user_reviews['item_id'].astype(int)
```

#### **Columna "recommend**

```python
# cambiar columna a tipo bool

user_reviews['recommend']=user_reviews['recommend'].astype(bool)
```

### **Feature Engineering**

A continuación se llevará a cabo un analisis de sentimiento a la columna "review", mediante este proceso se obtendra una nueva columna que llamaremos "sentiment_analysis" la misma contrendra valores 0 , 1 , 2 . Estos valores se obtendran en base a la polaridad obtenida por ele analisis de sentimiento, donde para aquellos comentarios considereados negativos se adoptara un valor 0, para comentarios considerados neutros se adoptará un valor de 1 y para comentarios considerados positivos se adoptara el valor 2.

```python
# Funcion que efectua un analisis de sentimiento sobre la columna de datos introducida como parametro 

def analisis_sentimiento(texto):
    analisis=TextBlob(texto)
    if analisis.polarity>0:
        return 2
    elif analisis.polarity<0:
        return 0
    elif analisis.polarity==0:
        return 1

# Crea una nueva columna llamada "sentiment_analysis" obtenida mediante la aplicacion del analisis de sentimiento sobre la columna "review"
user_reviews['sentiment_analysis']=user_reviews['review'].apply(analisis_sentimiento)
user_reviews.head(5)
```

```python
# Elimino la columna review ya que no será útil para el entrenamiento del modelo de machine learning que se efecutará mas adelante

user_reviews.drop('review',axis=1,inplace=True) 
user_reviews.head(3)
```
